```python
# ==========================================
# üß† VGG16 Binary Classification (1300 images)
# ==========================================

# --- 1Ô∏è‚É£ Mount Drive ---
from google.colab import drive
drive.mount('/content/drive')

# --- 2Ô∏è‚É£ Imports ---
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import VGG16
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
import matplotlib.pyplot as plt
import os, shutil
from sklearn.model_selection import train_test_split

# --- 3Ô∏è‚É£ Paths ---
dataset_dir = "/content/drive/MyDrive/Mobile_Dataset"
combined_dir = dataset_dir + "/all_images"

# --- 4Ô∏è‚É£ Combine old train/test if needed ---
train_dir = os.path.join(dataset_dir, "train")
test_dir = os.path.join(dataset_dir, "test")
os.makedirs(combined_dir, exist_ok=True)

for split_dir in [train_dir, test_dir]:
    for class_name in os.listdir(split_dir):
        class_path = os.path.join(split_dir, class_name)
        target_path = os.path.join(combined_dir, class_name)
        os.makedirs(target_path, exist_ok=True)
        for img in os.listdir(class_path):
            src = os.path.join(class_path, img)
            dst = os.path.join(target_path, img)
            if not os.path.exists(dst):
                shutil.copy(src, dst)

print("‚úÖ Combined dataset created at:", combined_dir)

# --- 5Ô∏è‚É£ Split 70/15/15 ---
base_split_dir = "/content/split_dataset"
train_split = os.path.join(base_split_dir, "train")
val_split   = os.path.join(base_split_dir, "val")
test_split  = os.path.join(base_split_dir, "test")

for d in [train_split, val_split, test_split]:
    os.makedirs(d, exist_ok=True)

for class_name in os.listdir(combined_dir):
    class_path = os.path.join(combined_dir, class_name)
    imgs = os.listdir(class_path)

    train_imgs, temp_imgs = train_test_split(imgs, test_size=0.30, random_state=42)
    val_imgs, test_imgs = train_test_split(temp_imgs, test_size=0.50, random_state=42)

    for split, img_list in zip([train_split, val_split, test_split], [train_imgs, val_imgs, test_imgs]):
        class_target = os.path.join(split, class_name)
        os.makedirs(class_target, exist_ok=True)
        for img in img_list:
            shutil.copy(os.path.join(class_path, img), os.path.join(class_target, img))

print("‚úÖ Dataset split 70/15/15 complete")

# --- 6Ô∏è‚É£ Data Generators ---
img_height, img_width = 224, 224
batch_size = 32

train_gen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)
test_gen = ImageDataGenerator(rescale=1./255)

train_data = train_gen.flow_from_directory(train_split, target_size=(img_height, img_width),
                                           batch_size=batch_size, class_mode='binary')
val_data = test_gen.flow_from_directory(val_split, target_size=(img_height, img_width),
                                        batch_size=batch_size, class_mode='binary')
test_data = test_gen.flow_from_directory(test_split, target_size=(img_height, img_width),
                                         batch_size=batch_size, class_mode='binary')

# --- 7Ô∏è‚É£ Build Model ---
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(img_height, img_width, 3))
for layer in base_model.layers:
    layer.trainable = False

x = Flatten()(base_model.output)
x = Dense(512, activation='relu')(x)
x = Dropout(0.5)(x)
output = Dense(1, activation='sigmoid')(x)

model = Model(inputs=base_model.input, outputs=output)

model.compile(optimizer=Adam(learning_rate=1e-4),
              loss='binary_crossentropy',
              metrics=['accuracy'])

# --- 8Ô∏è‚É£ Stage 1: Train Top Layers ---
early_stop = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
checkpoint = ModelCheckpoint("/content/drive/MyDrive/vgg16_binary_best.h5",
                             monitor='val_accuracy', save_best_only=True, verbose=1)

print("üöÄ Stage 1: Training with frozen base...")
history1 = model.fit(train_data, validation_data=val_data,
                     epochs=10, callbacks=[early_stop, checkpoint])

# --- 9Ô∏è‚É£ Stage 2: Fine-Tuning ---
print("üîß Unfreezing last 6 conv layers for fine-tuning...")
for layer in base_model.layers[:-6]:
    layer.trainable = False
for layer in base_model.layers[-6:]:
    layer.trainable = True

model.compile(optimizer=Adam(learning_rate=1e-5),
              loss='binary_crossentropy',
              metrics=['accuracy'])

print("üöÄ Stage 2: Fine-tuning...")
history2 = model.fit(train_data, validation_data=val_data,
                     epochs=10, callbacks=[early_stop, checkpoint])

# --- üîü Evaluate on Test Set ---
loss, acc = model.evaluate(test_data)
print(f"‚úÖ Final Test Accuracy: {acc*100:.2f}%")

# --- 1Ô∏è‚É£1Ô∏è‚É£ Plot Results ---
plt.figure(figsize=(12,5))
acc = history1.history['accuracy'] + history2.history['accuracy']
val_acc = history1.history['val_accuracy'] + history2.history['val_accuracy']
losses = history1.history['loss'] + history2.history['loss']
val_losses = history1.history['val_loss'] + history2.history['val_loss']

plt.subplot(1,2,1)
plt.plot(acc, label='Train Accuracy')
plt.plot(val_acc, label='Validation Accuracy')
plt.title('Accuracy (Frozen + Fine-tuned)')
plt.legend()

plt.subplot(1,2,2)
plt.plot(losses, label='Train Loss')
plt.plot(val_losses, label='Validation Loss')
plt.title('Loss (Frozen + Fine-tuned)')
plt.legend()
plt.show()

print("‚úÖ Training complete. Best model saved to Drive.")
```
